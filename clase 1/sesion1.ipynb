```python
# Importar la librería GeoPandas para manejo de datos geoespaciales
import geopandas as gpd
```

```python
# Leer el archivo shapefile de departamentos de Colombia y almacenarlo en un GeoDataFrame
colombia = gpd.read_file('/content/drive/MyDrive/curso geoinformacion python/colombia/MGN_ANM_DPTOS.shp')
```

```python
# Mostrar el contenido completo del GeoDataFrame
colombia
```

```python
# Consultar las dimensiones del GeoDataFrame (filas, columnas)
colombia.shape
```

```python
# Listar los nombres de todas las columnas del GeoDataFrame
colombia.columns
```

```python
# Mostrar las primeras 5 filas del GeoDataFrame
colombia.head(2)
```

```python
# Mostrar las últimas 4 filas del GeoDataFrame
colombia.tail(2)
```

```python

# Consultar el sistema de referencia de coordenadas (CRS) del GeoDataFrame
colombia.crs
```

```python
# Contar el número de departamentos únicos en la columna DPTO_CNMBR
colombia["DPTO_CNMBR"].nunique()
```

```python
# Obtener un arreglo con los nombres únicos de todos los departamentos
colombia["DPTO_CNMBR"].unique()
```

```python
# Generar un mapa básico con todos los departamentos de Colombia
colombia.plot()
```

```python
# Generar un mapa coloreando cada departamento según su nombre
colombia.plot(column='DPTO_CNMBR')
```

```python
# Filtrar el GeoDataFrame para obtener solo el departamento del Meta
meta = colombia[colombia["DPTO_CNMBR"] == "META"]
meta.plot()
```

```python
# Graficar el mapa del departamento del Meta con tamaño por defecto
meta.plot()
```

```python
# Graficar el mapa del Meta con un tamaño de figura de 10x10 pulgadas
meta.plot(figsize=(10,10))
```

```python
# Graficar el Meta coloreado por nombre de departamento, con leyenda y tamaño 12x10
meta.plot(column='DPTO_CNMBR', legend=True, figsize=(10,10))
```

```python
# Graficar el Meta con relleno blanco, bordes azules y tamaño 10x10
meta.plot(color='green', edgecolor='blue', figsize=(10,10))
```

```python
# Exportar el GeoDataFrame del Meta a un nuevo archivo shapefile
meta.to_file('/content/drive/MyDrive/curso geoinformacion python/meta/meta.shp')
```

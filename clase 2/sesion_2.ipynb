```python
import geopandas as gpd

# Ruta del archivo shapefile
shapefile_path = "/content/drive/MyDrive/CURSOS/curso geoinformacion python/CLASE 2/colombia_dpto/MGN_ANM_DPTOS.shp"

# Cargar el archivo shapefile
departamentos = gpd.read_file(shapefile_path)
print(f"Shapefile '{shapefile_path}' cargado.")
```

```python
if 'departamentos' in locals() and 'DPTO_CNMBR' in departamentos.columns:
    num_categorias = departamentos['DPTO_CNMBR'].nunique()
    categorias = departamentos['DPTO_CNMBR'].unique()

    print(f"Número de categorías únicas en la columna 'DPTO_CNMBR': {num_categorias}")
    print("\nCategorías en la columna 'DPTO_CNMBR':")
    for categoria in sorted(categorias):
        print(f"- {categoria}")
else:
    print("Asegúrate de que el GeoDataFrame 'departamentos' esté cargado y contenga la columna 'DPTO_CNMBR'.")
```

```python
import matplotlib.pyplot as plt

# Visualizar la geodata de departamentos con color verde oscuro y sin ejes
departamentos.plot(figsize=(10, 10), color='darkgreen')
plt.title('Segunda Sesión', color='red')
plt.axis('off') # Quita los ejes (y por ende los márgenes y etiquetas)
plt.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
import os
import pandas as pd # Import pandas for data type checking

# Asegurarse de que el GeoDataFrame del Meta y las estaciones estén disponibles
# Si no lo estuvieran, se cargarían:
# meta_shapefile_path = '/content/drive/MyDrive/CURSOS/curso geoinformacion python/CLASE 2/dpto_meta/META.shp'
# departamento_meta_cargado = gpd.read_file(meta_shapefile_path)
# estaciones_shapefile_path = "/content/drive/MyDrive/CURSOS/curso geoinformacion python/CNE_IDEAM/CNE_IDEAM_Final.shp"
# estaciones_ideam = gpd.read_file(estaciones_shapefile_path)

# Realizar el 'clip' de los puntos de estaciones con el polígono del Meta
estaciones_meta = gpd.clip(estaciones_ideam, departamento_meta_cargado).copy()

print(f"Se han recortado {len(estaciones_meta)} estaciones dentro del departamento del Meta.")

# Visualizar el corte
fig, ax = plt.subplots(1, 1, figsize=(12, 12))

# Plotear el departamento del Meta como capa base
departamento_meta_cargado.plot(ax=ax, color='lightgray', edgecolor='black', linewidth=0.8)

# Plotear las estaciones recortadas sobre el mapa del Meta
estaciones_meta.plot(ax=ax, marker='x', color='blue', markersize=20, label='Estaciones en Meta')

# Añadir un título
ax.set_title('Estaciones Climáticas en el Departamento del Meta', fontsize=15)

# Quitar los ejes
ax.set_axis_off()

# Añadir leyenda
plt.legend()

plt.show()

# Exportar el nuevo shapefile a la carpeta "estaciones_meta"
output_base_path = "/content/drive/MyDrive/CURSOS/curso geoinformacion python/CLASE 2"
output_folder_meta_stations = os.path.join(output_base_path, "estaciones_meta")

# Crear la carpeta si no existe
if not os.path.exists(output_folder_meta_stations):
    os.makedirs(output_folder_meta_stations)
    print(f"Carpeta '{output_folder_meta_stations}' creada exitosamente.")
else:
    print(f"La carpeta '{output_folder_meta_stations}' ya existe.")

# Definir la ruta completa para el nuevo shapefile de estaciones
output_shapefile_meta_stations = os.path.join(output_folder_meta_stations, "estaciones_meta.shp")

# Convertir columnas de tipo datetime a string para evitar RuntimeWarning al exportar a shapefile
# El formato Shapefile no tiene un tipo de dato nativo para DateTime.
for col in ['FECHA_INST', 'FECHA_SUSP']:
    if col in estaciones_meta.columns and pd.api.types.is_datetime64_any_dtype(estaciones_meta[col]):
        estaciones_meta[col] = estaciones_meta[col].dt.strftime('%Y-%m-%d') # Formato de fecha estándar

# Exportar el GeoDataFrame recortado a un nuevo shapefile
estaciones_meta.to_file(output_shapefile_meta_stations)

print(f"Shapefile de estaciones del Meta exportado a: {output_shapefile_meta_stations}")
```

```python
import matplotlib.pyplot as plt

# Definir los departamentos que pertenecen a la región de la Orinoquía
departamentos_orinoquia_nombres = ['ARAUCA', 'CASANARE', 'META', 'VICHADA']

# Filtrar el GeoDataFrame original para incluir solo los departamentos de la Orinoquía
departamentos_orinoquia = departamentos[departamentos['DPTO_CNMBR'].isin(departamentos_orinoquia_nombres)]

# Visualizar los departamentos de la Orinoquía
fig, ax = plt.subplots(1, 1, figsize=(12, 12))

departamentos_orinoquia.plot(ax=ax, color='skyblue', edgecolor='black', linewidth=0.8)

# Añadir un título
ax.set_title('Departamentos de la Orinoquía Colombiana', fontsize=15)

# Quitar los ejes para una visualización más limpia
ax.set_axis_off()

plt.show()

print(f"Departamentos de la Orinoquía visualizados: {departamentos_orinoquia_nombres}")
```

```python
import matplotlib.pyplot as plt

# Disolver los departamentos de la Orinoquía en una sola región
# Por defecto, dissolve mantiene las columnas con valores únicos o toma el primer valor
# Si quieres agregar un nombre de región, puedes hacerlo después
orinoquia_dissolved = departamentos_orinoquia.dissolve(by=None) # by=None disuelve todas las geometrías en una sola

# Asignar un nombre a la nueva región
orinoquia_dissolved['DPTO_CNMBR'] = 'Orinoquia'

print("Región de Orinoquia creada por disolución.")

# Visualizar la nueva región disuelta
fig, ax = plt.subplots(1, 1, figsize=(12, 12))

orinoquia_dissolved.plot(ax=ax, color='lightcoral', edgecolor='black', linewidth=0.8)

# Añadir un título
ax.set_title('Región de Orinoquía (Departamentos Disueltos)', fontsize=15)

# Quitar los ejes para una visualización más limpia
ax.set_axis_off()

plt.show()
```
